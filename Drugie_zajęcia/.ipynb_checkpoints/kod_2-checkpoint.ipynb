```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("Lab2-Transactions")
    .getOrCreate()
)
spark.sparkContext.setLogLevel("WARN")
print(f"Spark {spark.version} — gotowy")
```

```python
spark
```

```python
df = spark.read.json("transactions_10k.jsonl")

print(f"Liczba rekordów: {df.count()}")
df.printSchema()
```

```python
df.show(10, truncate=False)
```

```python
from pyspark.sql.functions import to_timestamp, col

df = df.withColumn("timestamp", to_timestamp(col("timestamp"), "yyyy-MM-dd HH:mm:ss"))

df.printSchema()  # timestamp powinien być teraz 'timestamp (nullable = true)'
```

```python
from pyspark.sql.functions import count, sum as _sum, avg, round as _round

store_summary = (
    df.groupBy("store")
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
        _round(avg("amount"), 2).alias("srednia_PLN"),
    )
    .orderBy("store")
)
store_summary.show()
```

```python
from pyspark.sql.functions import min as _min, max as _max

df.groupBy("category") \
  .agg(
      _min("amount").alias("min_PLN"),
      _max("amount").alias("max_PLN")
  ) \
  .orderBy("category") \
  .show()
```

```python
from pyspark.sql.functions import window

hourly = (
    df.groupBy(window("timestamp", "1 hour"))    # okno 1-godzinne
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
    )
    .orderBy("window")
)
hourly.show(truncate=False)
```

```python
(
    hourly
    .select(
        col("window.start").alias("od"),
        col("window.end").alias("do"),
        "liczba_tx",
        "suma_PLN",
    )
    .show(truncate=False)
)
```

```python
hourly.printSchema()
```

```python
from pyspark.sql.functions import window, count, sum as _sum, round as _round, col

hourly_store = (
    df.groupBy(
        window("timestamp", "30 minutes"),
        "store"
    )
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN")
    )
    .orderBy(col("window.start"), col("store"))
)

(
    hourly_store
    .select(
        col("window.start").alias("od"),
        col("window.end").alias("do"),
        "store",
        "liczba_tx",
        "suma_PLN"
    )
    .show(truncate=False)
)
```

```python
from pyspark.sql.functions import window, sum as _sum, round as _round, col, desc

krakow_best_hour = (
    df.filter(col("store") == "Kraków")
    .groupBy(window("timestamp", "1 hour"))
    .agg(
        _round(_sum("amount"), 2).alias("suma_PLN")
    )
    .orderBy(desc("suma_PLN"))
)

(
    krakow_best_hour
    .select(
        col("window.start").alias("od"),
        col("window.end").alias("do"),
        "suma_PLN"
    )
    .show(truncate=False)
)
```

```python
sliding = (
    df.groupBy(window("timestamp", "1 hour", "30 minutes"))  # szerokość 1h, krok 30min
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
    )
    .select(
        col("window.start").alias("od"),
        col("window.end").alias("do"),
        "liczba_tx",
        "suma_PLN",
    )
    .orderBy("od")
)
sliding.show(truncate=False)
```

```python
tumbling_rows = (
    df.groupBy(window("timestamp", "1 hour"))
    .agg(count("tx_id"))
    .count()
)
sliding_rows = (
    df.groupBy(window("timestamp", "1 hour", "30 minutes"))
    .agg(count("tx_id"))
    .count()
)
print(f"Tumbling (1h):          {tumbling_rows} okien")
print(f"Sliding  (1h / 30min):  {sliding_rows} okien")

```

```python
from pyspark.sql.functions import window, avg, round as _round, col, asc

gdansk_lowest_avg = (
    df.filter(col("store") == "Gdańsk")
    .groupBy(window("timestamp", "1 hour"))
    .agg(
        _round(avg("amount"), 2).alias("srednia_PLN")
    )
    .orderBy(asc("srednia_PLN"))
)

(
    gdansk_lowest_avg
    .select(
        col("window.start").alias("od"),
        col("window.end").alias("do"),
        "srednia_PLN"
    )
    .show(truncate=False)
)
```

```python
from pyspark.sql.functions import window, count, col

category_0900_0930 = (
    df.groupBy(
        window("timestamp", "30 minutes"),
        "category"
    )
    .agg(
        count("tx_id").alias("liczba_tx")
    )
    .filter(
        (col("window.start").cast("string").contains("09:00:00")) &
        (col("window.end").cast("string").contains("09:30:00"))
    )
    .orderBy("category")
)

(
    category_0900_0930
    .select(
        col("window.start").alias("od"),
        col("window.end").alias("do"),
        "category",
        "liczba_tx"
    )
    .show(truncate=False)
)
```

```python
from pyspark.sql.functions import window, count, col, desc

peak_15min = (
    df.groupBy(window("timestamp", "15 minutes"))
    .agg(
        count("tx_id").alias("liczba_tx")
    )
    .orderBy(desc("liczba_tx"))
)

(
    peak_15min
    .select(
        col("window.start").alias("od"),
        col("window.end").alias("do"),
        "liczba_tx"
    )
    .show(truncate=False)
)
```
